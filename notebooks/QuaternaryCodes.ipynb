# Quaternary Codes

Enumerating and sorting all quaternary codes (formerly known as "Kirk sequences") up to trees with 7 leaf nodes

```python
%load_ext autoreload
%autoreload 2
import numpy as np
import matplotlib.pyplot as plt
import sys
sys.path.append("..")
from binarytrees import *
```

```python
plt.figure(figsize=(8, 8))
trees = []
for n in range(2, 7):
    for T in enumerate_trees(n):
        seq = T.get_quaternary_code()
        binary = "".join(["%i"%i for i in seq]) # Convert to binary string
        decimal = int(binary, 2)
        trees.append((T, n, binary, decimal))
trees = sorted(trees, key = lambda x: x[-1]) # Sort in ascending decimal order

sequence = np.array([t[-1] for t in trees])
print(",".join(["%i"%s for s in sequence]))
print(sequence[1::] - sequence[0:-1])

for i, (T, n, binary, decimal) in enumerate(trees):
    plt.clf()
    T.draw()
    plt.title("{} ({}), {} leaves".format(binary, decimal, n+1))
    plt.savefig("Tree{}.png".format(i+1))
```
